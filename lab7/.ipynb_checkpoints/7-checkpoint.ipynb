```python
from cProfile import label
import numpy as np
import matplotlib.pyplot as plt
import math

month=np.arange(1,10)
days =np.array([31,29,31,30,31,30,31,31,30])
sum_sun=np.zeros(9,dtype=float)
mean_sun=np.zeros(9,dtype=float)
with open('daily_KP_SUN_2020.csv','r',encoding='utf-8') as f1:
    for i in range(3):
        f1.readline()
    for lines in f1:
        line=lines.strip()
        elements=line.split(',')
        sum_sun[int(elements[1])-1]+=float(elements[3])
    for i in range(9):
        mean_sun[i]=sum_sun[i]/days[i]
    
    plt.subplot(211)
    plt.title("Total Bright Sunshine - Hong Kong King's Park")
    plt.xlabel('Month')
    plt.ylabel('Sum_hours')
    plt.bar(month,sum_sun)
    plt.subplot(212)
    plt.title("Average Bright Sunshine - Hong Kong King's Park")
    plt.xlabel('Month')
    plt.ylabel('Mean_hours')
    plt.bar(month,mean_sun,color='red')
    plt.subplots_adjust(hspace=0.6) # 调整子图之间的纵向间隔，避免上方子图的横坐标标签与下方子图的标题重叠
    plt.show()
```
